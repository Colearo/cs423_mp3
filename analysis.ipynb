```python
import plotly.offline as py
import plotly.graph_objs as go

major_faults = []
minor_faults = []
sum_maj = 0
sum_min = 0

with open("./profile1.data") as f :
    jiffies = []
    for line in f.readlines() :
        try :
            s = [int(x) for x in line.strip().split(" ")]
        except :
            break
        if s[0] == 0 :
            break
        sum_maj += s[2]
        sum_min += s[1]
        jiffies.append(s[0])
        major_faults.append(sum_maj)
        minor_faults.append(sum_min)
    print("Completion time", jiffies[-1] - jiffies[0])

traces = []

trace = go.Scatter(
    x = [x*0.05 for x in range(len(minor_faults))],
    y = [y for y in minor_faults],
    mode = 'lines',
    name = "minor faults count"
)
traces.append(trace)

trace = go.Scatter(
    x = [x*0.05 for x in range(len(major_faults))],
    y = [y for y in major_faults],
    mode = 'lines',
    name = "major faults count"
)
traces.append(trace)

py.init_notebook_mode(connected=True)
py.iplot({
    "data": traces,
    "layout": dict(title = 'Page Fault Count Running Two Workload with Both Random Access',
              xaxis = dict(title = 'Time/s'),
              yaxis = dict(title = 'Page Fault Count'),
              )
})
```

```python
import plotly.offline as py
import plotly.graph_objs as go

utilization = dict()
page_fault = dict()

for i, j in [(1,3),(5,4),(11,5)]:
    utilization[i] = 0
    page_fault[i] = []
    jiffies = []
    sum_maj = 0
    sum_min = 0
    with open("./profile%d.data" % j) as f :
        for line in f.readlines() :
            try :
                s = [int(x) for x in line.strip().split(" ")]
            except :
                break
            jiffies.append(s[0])
            utilization[i] += s[3]
            sum_maj += s[2]
            sum_min += s[1]
            page_fault[i].append(sum_maj + sum_min)
    print("Instance %d completion time: %d " % (i, jiffies[-1] - jiffies[0]))
    print("Total page faults is: major %d, minor %d" % (sum_maj, sum_min));
    utilization[i] = utilization[i]/(jiffies[-1] - jiffies[0])
traces = []

trace = go.Scatter(
    x = [x for x in sorted(utilization)],
    y = [utilization[x] for x in sorted(utilization)],
    mode = 'lines+markers+text',
    name = "utilization time",
    fill = 'tonexty',
    line = {'shape': 'spline', 'smoothing': 0.4},
    text = ["%.3f" % utilization[x] for x in sorted(utilization)],
    textposition = 'top center',
)
traces.append(trace)

py.init_notebook_mode(connected=True)
py.iplot({
    "data": traces,
    "layout": dict(title = 'CPU Total Utilization of All N Instances of Work Process',
              xaxis = dict(title = 'Instances / # of Processes', nticks = 3, range = [0, 12]),
              yaxis = dict(title = 'CPU Utilization'),
            )
})

traces = []
for i in sorted(page_fault.keys()) :
    trace = go.Scatter(
        x = [x*0.05 for x in range(len(page_fault[i]))],
        y = [y for y in page_fault[i]],
        mode = 'lines',
        name = "%d instances faults count" % i
    )
    traces.append(trace)

py.init_notebook_mode(connected=True)
py.iplot({
    "data": traces,
    "layout": dict(title = 'Page Fault Count Running N Processes',
              xaxis = dict(title = 'Time/s'),
              yaxis = dict(title = 'Page Fault Count'),
              )
})
```
