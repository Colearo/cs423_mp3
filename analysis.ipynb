```python
import plotly.offline as py
import plotly.graph_objs as go

major_faults = []
minor_faults = []

with open("./profile1.data") as f :
    for line in f.readlines() :
        s = [int(x) for x in line.strip().split(" ")]
        if s[0] == 0 :
            break
        print(s)
        major_faults.append(s[2])
        minor_faults.append(s[1])

traces = []

trace = go.Scatter(
    x = [x for x in range(len(minor_faults))],
    y = [y for y in minor_faults],
    mode = 'lines',
    name = "minor faults count"
)
traces.append(trace)

trace = go.Scatter(
    x = [x for x in range(len(major_faults))],
    y = [y for y in major_faults],
    mode = 'lines',
    name = "major faults count"
)
traces.append(trace)

py.init_notebook_mode(connected=True)
py.iplot({
    "data": traces,
    "layout": dict(title = 'Page Fault Count Running Two Workload',
              xaxis = dict(title = 'one time per 50 ms'),
              yaxis = dict(title = 'Page Fault Count'),
              )
})
```

```python
import plotly.offline as py
import plotly.graph_objs as go

utilization = dict()

for i, j in {(1,3),(5,4),(11,5)}:
    utilization[i] = 0
    with open("./profile%d.data" % j) as f :
        for line in f.readlines() :
            s = [int(x) for x in line.strip().split(" ")]
            if s[0] == 0 :
                break
            print(s)
            utilization[i] += s[3]

traces = []

trace = go.Scatter(
    x = [x for x in sorted(utilization)],
    y = [utilization[x] for x in sorted(utilization)],
    mode = 'lines',
    name = "utilization time",
    fill = 'tonexty',
)
traces.append(trace)

py.init_notebook_mode(connected=True)
py.iplot({
    "data": traces,
    "layout": dict(title = 'CPU Total Utilization of All N Instances of Work Process',
              xaxis = dict(title = 'Instances', nticks = 3, zeroline = True),
              yaxis = dict(title = 'Utilization Time'),
            )
})
```
